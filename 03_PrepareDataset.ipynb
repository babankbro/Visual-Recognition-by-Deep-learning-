```python
dataset_path = "../Dataset/"
Xs = np.load(dataset_path+ "ThaiTest_X_128.npy")
Ys = np.load(dataset_path+ "ThaiTest_Y_128.npy")
print(Xs.shape, Ys.shape)
```

```python
Xs[[0, 1000, 2000]].shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedKFold
```

```python
seed = 30
Number_Of_Fold = 10

skf = StratifiedKFold(n_splits=Number_Of_Fold, random_state=seed)
```

```python
i = 1
train_test_indexes = []
for train_index, test_index in skf.split(Xs, Ys):
    print(i , Ys[test_index].shape, test_index[:20])
    
    train_test_indexes.append( { 'train':train_index, "test": test_index } ) 
    
    i += 1
```

```python
train_index =train_test_indexes[9]['train']
test_index =train_test_indexes[9]['test']
print(test_index.shape, test_index)
```

```python
for i in range(5):
    train_index =train_test_indexes[i]['train']
    test_index =train_test_indexes[i]['test']
    
    
    X_test = Xs[test_index]
    Y_test = Ys[test_index]
    
    X_train = Xs[train_index]
    Y_train = Ys[train_index]
    
    
    print(i, "Train:", Y_train.shape, "Test:", Y_test.shape, )
    train_index, val_index, y_train, y_val = train_test_split(train_index, Ys[train_index], 
                                                                  test_size = 1/9, random_state=seed)
    X_train = Xs[train_index]
    Y_train = Ys[train_index]
    
    X_val = Xs[val_index]
    Y_val = Ys[val_index]
    
    
    print(i, "Train:", Y_train.shape, "Val:", Y_val.shape, "Test:", Y_test.shape )
    np.save(dataset_path+ "Train_X_128_Set"+str(i+1)+".npy", X_train)
    np.save(dataset_path+ "Train_Y_128_Set"+str(i+1)+".npy", Y_train)
    
    np.save(dataset_path+ "Val_X_128_Set"+str(i+1)+".npy", X_val)
    np.save(dataset_path+ "Val_Y_128_Set"+str(i+1)+".npy", Y_val)
    
    np.save(dataset_path+ "Test_X_128_Set"+str(i+1)+".npy", X_test)
    np.save(dataset_path+ "Test_Y_128_Set"+str(i+1)+".npy", Y_test)
    
    
```
